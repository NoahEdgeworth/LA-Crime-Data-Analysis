```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Crime_Data_from_2020_to_Present.csv')
df.head()
```

```python
focus_columns = ['DR_NO','Date Rptd','DATE OCC','TIME OCC','AREA','AREA NAME','Part 1-2','Crm Cd','Crm Cd Desc','Vict Age','Vict Sex',
                 'Vict Descent','Premis Desc','Weapon Desc','Status Desc','LOCATION','LAT','LON']
new_df = df[focus_columns]
```

```python
new_df.isnull().sum()
```

```python
new_df.head()
```

```python
# Removing any ages that are less than 0 for victims, and replacing with NULL
new_df.loc[new_df['Vict Age'] < 0, 'Vict Age'] = np.nan
```

```python
# Drop all rows with no LAT or LON value
new_df = new_df[(new_df['LAT'] != 0) & (new_df['LON'] != 0)]
```

```python
# Saving this pandas table to a CSV now that it is cleaned, so we can move on to diving a little deeper into it
new_df.to_csv('Crime_Data_from_2020_to_Present_CLEANED.csv', index=False)
```
